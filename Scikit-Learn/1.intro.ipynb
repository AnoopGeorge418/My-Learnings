```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns; sns.set()
```

```python
N = 30
x =np.arange(N)
y = 4 * x - 7 + (5 * np.random.randn(N))
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
yfit = model.predict(x[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x, yfit, color = 'red')
```

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=500, centers=3, cluster_std=1.2)

plt.scatter(x[:, 0], x[:, 1], c=y, cmap='autumn')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Generated Blobs')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.svm import SVC
```

```python
mdl = SVC(gamma='auto')
mdl.fit(X_train, y_train)
```

```python
y_pred = mdl.predict(X_test)
```

```python
acc = 100 * np.sum(y_pred == y_test) / y_test.size
print('Accuracy is: ', acc, '%')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mdl = RandomForestClassifier(n_estimators=20)
mdl.fit(X_train, y_train)
```

```python
y_pred = mdl.predict(X_test)
acc = 100 * np.sum(y_pred == y_test) / y_test.size
print(f'Accuracy is: {acc}%')
```
