```python
sns.set()
x = pd.Series(50 * np.random.randn(5000))
y = pd.Series(200 * np.random.randn(5000))
z = pd.Series(100 * np.random.randn(5000) + 500)
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})
# data.head()
for col in data.columns:
    plt.hist(data[col], density = True, alpha = 0.5)
    sns.kdeplot(data[col], shade = True)
```

```python
for col in data.columns:
    sns.distplot(data[col])
```

```python
sns.set(style='whitegrid')
sns.jointplot(x = 'X', y = 'Y', data = data, kind = 'kde')
```

```python
sns.pairplot(data=data)
```

```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'X': np.random.normal(0, 2, 1000), 
                  'Y': np.random.normal(3, 5, 1000), 
                  'Z': np.random.normal(-4, 10, 1000)})
```

```python
df.head()
```

```python
sns.set()
plt.figure(figsize=(12, 6))
for col in df.columns:
    plt.hist(df[col], density = True, alpha = 0.5, bins = 30)
    sns.kdeplot(df[col], shade = True)
```

```python
for col in df.columns:
    sns.distplot(df[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = 'X', y = 'Y', data = df, kind='kde')
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width', data = iris,
            hue='species', style='species')
```

```python
sns.relplot(x = 'sepal_length', y = 'petal_length', data = iris,
            hue='petal_length', style='species')
```

```python
iris.head()
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width', data = iris,
            kind='line', ci = None, hue='species', style = 'species')
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width', data = iris,
            dashes=False,kind='line', markers=True, ci = None, hue='species', style = 'species')
```

```python
sns.relplot(x = 'sepal_length', y = 'sepal_width', data = iris,
            dashes=True, kind='line', markers=True, ci = None, hue='species', style = 'species')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.time.unique()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue='smoker', style='smoker', col='time', data=tips)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
len(fmri.subject.unique())
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue='event', 
            style='event', col='subject', col_wrap=5, 
            kind='line', data=fmri.query("region == 'frontal'"))
```

```python
sns.set(style='ticks', color_codes=True)
sns.catplot(data = tips, x = 'day', y= 'total_bill')
```

```python
sns.set(style='ticks', color_codes=True)
sns.catplot(data = tips, x = 'day', y= 'total_bill', jitter = False)
```

```python
sns.set(style='ticks', color_codes=True)
sns.catplot(data = tips, x = 'day', y= 'total_bill',kind='boxen', hue='sex')
```

```python
sns.catplot(data = tips, x = 'day', y= 'total_bill',kind='violin', hue='sex')
```

```python
sns.catplot(data = tips, x = 'day', y= 'total_bill',kind='bar', hue='sex')
```
