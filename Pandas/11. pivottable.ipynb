```python
import numpy as np 
import pandas as pd
```

```python
from seaborn import load_dataset
```

```python
T = load_dataset('titanic')
T.head()
```

```python
T.rename(columns={'sex': 'gender'}, inplace = True)
T
```

```python
T.deck.isnull().sum()
```

```python
T.drop(['deck'], axis=1, inplace=True)
```

```python
T
```

```python
T.groupby(['gender', 'class'])['survived'].mean().unstack()
```

```python
T.pivot_table('survived', index = 'gender', columns='class', aggfunc='mean')
```

```python
agePartition = pd.cut(T['age'], [0, 18, 80])
```

```python
agePartition
```

```python
T.pivot_table('survived', ['gender', agePartition], 'class')
```
