```python
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('tableau-colorblind10')
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(111, projection='3d')
theta = np.linspace(-4*np.pi, 4*np.pi, 100)
Z = np.linspace(-2, 2, 100)
r = Z ** 2 + 1
X = r * np.sin(theta)
Y = r * np.cos(theta)
ax.plot(X, Y, Z, label = '3D Curve', color ='red')
plt.cm.inferno
plt.legend()
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('tableau-colorblind10')
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(111, projection='3d')
X = np.random.normal(0, 2, 100)
Y = np.random.normal(2, 4, 100)
Z = np.random.normal(6, 1, 100)
ax.scatter(X, Y, Z, c='b', marker = '^')
ax = fig.add_subplot(111, projection='3d', alpha = 0.7)
X = np.random.normal(10, 2, 100)
Y = np.random.normal(12, 4, 100)
Z = np.random.normal(16, 1, 100)
ax.scatter(X, Y, Z, c='r', marker = 's', alpha = 0.7)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm)
```
