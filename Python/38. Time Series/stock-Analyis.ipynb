```python
# Time Resampling - Year and frequency
df.head()
```

```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').min()['Open'].plot()
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
# quaterly End frequency
df.resample(rule='QS').min()['High'].plot()
```

```python
# business End frequency
df.resample(rule='BQS').min()['High'].plot()
```

```python
df['Open'].resample(rule = 'M').max().plot(kind = 'bar', figsize = (12, 6))
```

```python
df['Open'].plot(figsize = (12, 6))
```

```python
# Simple Moving Average
df['Open: 10 days rolling'] = df['Open'].rolling(10, 1).mean()
```

```python
df[['Open', 'Open: 10 days rolling']].plot(figsize = (12, 6))
```
