```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
print(iris['target'])
```

```python
X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

# Decesion Tree

```python
from sklearn.tree import DecisionTreeClassifier

tc = DecisionTreeClassifier()

tc.fit(X_train, y_train)

from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(tc, filled=True)
```

```python
y_pred = tc.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Post pruning

```python
from sklearn.tree import DecisionTreeClassifier

tc = DecisionTreeClassifier(max_depth=2)

tc.fit(X_train, y_train)

from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(tc, filled=True)
```

```python
y_pred = tc.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# PrePruning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
params
```

# PrePruning and Hyperparameter

```python
from sklearn.model_selection import GridSearchCV

treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel, param_grid = params, cv = 5, scoring = 'accuracy')

import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

# Descesion Tree Regressor

```python
from sklearn.datasets import load_diabetes

df = load_diabetes()
```

```python
df
```

```python
print(df.DESCR)
```

```python
X = pd.DataFrame(df['data'], columns=df['feature_names'])
```

```python
X
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
X_train.corr()
```

```python
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor

tr = DecisionTreeRegressor()
tr.fit(X_train, y_train)
```

# Hyper parameter tuning

```python
params = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5, 10, 15, 20, 25],
    'max_features': ['auto', 'sqrt', 'log2']
} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(tr, param_grid = params, cv = 5, scoring = 'neg_mean_squared_error')

import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
selectmodels = DecisionTreeRegressor(criterion='poisson', max_depth=3, max_features='sqrt', splitter='best')

selectmodels.fit(X_train, y_train)

plt.figure(figsize=(15, 10))
tree.plot_tree(selectmodels, filled=True)
plt.show()
```
