```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
%matplotlib inline
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
y_pred
```

```python
log.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, recall_score
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cl_rep = classification_report(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(score)
print(cm)
print(cl_rep)
print(recall)
```

# HyperParameter Tuninig And Cross Validation

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver)
```

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv = cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cl_rep = classification_report(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(score)
print(cm)
print(cl_rep)
print(recall)
```

### RandamizedSearch CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
random_cv = RandomizedSearchCV(estimator=model, param_distributions=params, cv = 5, scoring='accuracy')
```

```python
random_cv.fit(X_train, y_train)
```

```python
random_cv.best_score_
```

```python
random_cv.best_params_
```

```python
y_pred = random_cv.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cl_rep = classification_report(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(score)
print(cm)
print(cl_rep)
print(recall)
```

# ImBalanced Dataset

```python
from collections import Counter
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_clusters_per_class=1, random_state=10, n_redundant=0, weights=[0.99])
```

```python
X
```

```python
Counter(y)
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
class_weight = [{0:w, 1:y} for w in [1, 10, 50, 100] for y in [1, 10, 50, 100]]
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver, class_weight=class_weight)
```

```python
params
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv = cv, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cl_rep = classification_report(y_test, y_pred)

print(score)
print(cm)
print(cl_rep)
```

```python
sns.scatterplot(x = y_test, y = y_pred)
```

```python
sns.regplot(x = y_test, y = y_pred)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
from matplotlib import pyplot
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
dummy_model_prob = [0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model_prob = model.predict_proba(X_test)
```

```python
model_prob = model_prob[:, 1]
```

```python
dummy_model_aoc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)

print(dummy_model_aoc)
print(model_auc)
```

```python
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, threshold = roc_curve(y_test, model_prob)
```

```python
pyplot.plot(dummy_fpr, dummy_tpr, linestyle = '--', label = 'Dummy Model')
pyplot.plot(model_fpr, model_tpr, marker = '.', label = 'Logestic')

pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')

pyplot.legend()

pyplot.show()
```
